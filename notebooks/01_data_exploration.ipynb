## Dataset B Cleaning

```python
# Drop rows with null body or null answer
B_clean = B.copy()
print(f"Before: {B_clean.shape}")

B_clean = B_clean[B_clean["body"].notna()]
print(f"After dropping null body: {B_clean.shape}")

B_clean = B_clean[B_clean["answer"].notna()]
print(f"After dropping null answer: {B_clean.shape}")
```

```python
# Drop sparse/mostly-empty columns
cols_to_drop = ["version", "tag_5", "tag_6", "tag_7", "tag_8"]
B_clean = B_clean.drop(columns=cols_to_drop)
print(f"Remaining columns: {B_clean.columns.tolist()}")
```

```python
# Combine subject + body into a single "text" column
# subject has 5k nulls; we fill them with "" so we can concatenate.
# Format: "<subject>\n\n<body>". newline separator keeps them semantically distinct.
B_clean["subject"] = B_clean["subject"].fillna("")
B_clean["text"] = B_clean["subject"].str.strip() + "\n\n" + B_clean["body"].str.strip()
print(B_clean["text"].iloc[0])
```

```python
# S Drop duplicate rows based on the text column
before = len(B_clean)
B_clean = B_clean.drop_duplicates(subset=["text"])
print(f"Dropped {before - len(B_clean)} duplicate rows. Remaining: {len(B_clean)}")
```

```python
# Strip HTML tags from the text column. The tickets were originally web-scraped.

import re

def strip_html(text):
    text = re.sub(r"<[^>]+>", " ", text)       # replace tags with space
    text = re.sub(r"\s{3,}", "\n\n", text)      # collapse >2 consecutive whitespace to double newline
    return text.strip()

B_clean["text"] = B_clean["text"].apply(strip_html)
print("Sample after HTML strip:")
print(B_clean["text"].iloc[0])
```

```python
#Keep both English and German tickets 
B_clean = B_clean[B_clean["language"].isin(["en", "de"])]
print(f"Language distribution after filter:\n{B_clean['language'].value_counts()}")
```

## Dataset A Cleaning

```python
# Fix the b'...' bytes-literal prefix in the text column

import re

def fix_bytes_prefix(text):
    if isinstance(text, str) and re.match(r'^b[\'"]', text):
        # Remove the  b'  prefix and trailing quote
        inner = text[2:-1]
        # Unescape common escape sequences left behind by the repr
        inner = inner.replace("\\n", "\n").replace("\\t", "\t").replace("\\'", "'").replace('\\"', '"')
        return inner
    return text

A_clean = A.copy()
A_clean["text"] = A_clean["text"].apply(fix_bytes_prefix)

```

```python
print("=== Before fix (first 3) ===")
for t in A["text"].head(3):
    print(repr(t[:80]))

print("\n=== After fix (first 3) ===")
for t in A_clean["text"].head(3):
    print(repr(t[:80]))
```

```python
print("=== Sample of cleaned KB text ===\n")
for i, row in A_clean.head(3).iterrows():
    print(f"[{row['id']}] {row['title']}")
    print(row['text'][:300])
    print("---")
```
